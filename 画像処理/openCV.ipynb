```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('img_1.webp')
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
# BGRRGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# 
cv2.imwrite('./output.png',img)
```

```python
# 
resize_img = cv2.resize(img, (500,500))
plt.imshow(cv2.cvtColor(resize_img, cv2.COLOR_BGR2RGB))
```

```python
# 
h = round(img.shape[0]/2)
w = round(img.shape[1]/2)
small_img = cv2.resize(img, (w,h))
plt.imshow(cv2.cvtColor(small_img, cv2.COLOR_BGR2RGB))
```

```python
# 
trim_img = img[20:320,740:920]
plt.imshow(cv2.cvtColor(trim_img, cv2.COLOR_BGR2RGB))
```

```python
# 
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray_img, cv2.COLOR_BGR2RGB))
```

```python
# 
rotate_img = cv2.rotate(img, cv2.ROTATE_180)
plt.imshow(cv2.cvtColor(rotate_img, cv2.COLOR_BGR2RGB))
```

```python
# 
img_flip = cv2.flip(img,1)
plt.imshow(cv2.cvtColor(img_flip, cv2.COLOR_BGR2RGB))
```

```python
# 
img_2 = img[20:320,740:920]
h = img_2.shape[0]
w = img_2.shape[1]
img[0:h, 0:w] = img_2
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
